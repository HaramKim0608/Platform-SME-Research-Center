```python
comps_n=[]
date = []
sales = []
profit=[]
netin = []
asset = []
liab = []
capi = []
```

### 반복 작업 구간

```python
# 검색 불가 기업 발생 시
comps_n.append('')
date.append([])
sales.append([])
profit.append([])
netin.append([])
asset.append([])
liab.append([])
capi.append([])
```

기업 상세 조회 페이지 진입 후

```python
# 기업 상세 조회 페이지 진입 후
html = driver.page_source
soup = bs(html, 'html.parser')
```

```python
comp_n = soup.select_one('.table_title')
comp_n = comp_n.get_text()
comps_n.append(comp_n)
comps_n
```

```python
# 재무, 손익 정보 추출
comp_acc = soup.findAll('table', {'class': 'list_table type02'})
al=comp_acc[2].select('td')
```

```python
al3=[]
al4=[]
al5=[]
al6=[]
al7=[]
al8=[]
al9=[]

k=0
for i in al:
    if k%7 == 0:
        al2 = i.get_text()
        al3.insert(0, al2)
        k += 1
    elif k%7 == 1:
        al2 = i.get_text()
        al2_r = al2.replace(',','')
        al2_n = int(al2_r)
        al2_nn = al2_n/100000
        al4.insert(0, al2_nn) # 총자산
        k += 1
    elif k%7 == 2:
        al5.append(0) # 자본금
        k += 1
    elif k%7 == 3:
        al2 = i.get_text()
        al2_r = al2.replace(',','')
        al2_n = int(al2_r)
        al2_nn = al2_n/100000
        al6.insert(0, al2_nn) # 자본총계
        k += 1
    elif k%7 == 4:
        al2 = i.get_text()
        al2_r = al2.replace(',','')
        al2_n = int(al2_r)
        al2_nn = al2_n/100000
        al7.insert(0, al2_nn) # 매출액
        k += 1
    elif k%7 == 5:
        al2 = i.get_text()
        al2_r = al2.replace(',','')
        al2_n = int(al2_r)
        al2_nn = al2_n/100000
        al8.insert(0, al2_nn) # 영업이익
        k += 1
    else:
        al2 = i.get_text()
        al2_r = al2.replace(',','')
        al2_n = int(al2_r)
        al2_nn = al2_n/100000
        al9.insert(0, al2_nn) # 당기순이익
        k += 1
        
date.append(al3)
asset.append(al4)
liab.append(al5)
capi.append(al6)
sales.append(al7)
profit.append(al8)
netin.append(al9)
```

```python
# 확인용(마지막 입력값만 출력)
print(date[-1])
print(sales[-1])
print(profit[-1])
print(netin[-1])
print(asset[-1])
print(liab[-1])
print(capi[-1])
```

### 작업 후 Dataframe으로 합치는 과정

```python
date_df = pd.DataFrame(date, index=comps_n, columns = ['date', 'date', 'date', 'date', 'date'])
sales_df = pd.DataFrame(sales, index=comps_n, columns = ['sales', 'sales', 'sales', 'sales', 'sales'])
profit_df = pd.DataFrame(profit, index=comps_n, columns = ['profit', 'profit', 'profit', 'profit', 'profit'])
netin_df = pd.DataFrame(netin, index=comps_n, columns = ['netin', 'netin', 'netin', 'netin', 'netin'])
asset_df = pd.DataFrame(asset, index=comps_n, columns = ['asset', 'asset', 'asset', 'asset', 'asset'])
liab_df = pd.DataFrame(liab, index=comps_n, columns = ['liab', 'liab', 'liab', 'liab', 'liab'])
capi_df = pd.DataFrame(capi, index=comps_n, columns = ['capi', 'capi', 'capi', 'capi', 'capi'])
```

```python
comp_as_ni = pd.concat([date_df, sales_df, profit_df, netin_df, asset_df, liab_df, capi_df], axis=1)
```
