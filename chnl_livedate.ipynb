```python
# Crawler Package
import urllib.request as req
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Date and Time Package
from dateutil import rrule
from datetime import date
from datetime import datetime
from datetime import timedelta
from dateutil import relativedelta

# Data Preprocessing Package
import re
import pandas as pd
import os
import openpyxl
import random
import csv
```

```python
options = Options()

user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"
options.add_argument('user-agent=' + user_agent)

options.add_argument('--headless')
options.add_argument('--no-sadnbox')
options.add_argument('--disable-dev-shm-usage')

service = Service(executable_path='C:/Users/jrack/chromedriver.exe') # 크롬드라이버 저장 위치로 변경

options = webdriver.ChromeOptions()
options.add_experimental_option('excludeSwitches',['enable-logging'])

driver = webdriver.Chrome(service=service,options=options) # 크롬드라이버 실행. 아무것도 안 뜨는 게 정상임

#first='https://shoppinglive.naver.com/channels/'
```

```python
channel_product=[]
```

```python
last_height = driver.execute_script("return document.body.scrollHeight")
    
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") # 처음 화면 끝까지 스크롤
    time.sleep(1)
    driver.implicitly_wait(10)
    new_height = driver.execute_script("return document.body.scrollHeight") # 스크롤했을 때 높이 저장
    if new_height == last_height: # 스크롤한 창의 높이가 가장 하단 창의 높이와 같아지면 break
        break
    last_height = new_height

    
html = driver.page_source
soup = bs(html, 'html.parser')
comp=[]
channel_no = soup.find('a',{'class':'UnderlineItem_wrap_vmu1n UnderlineItem_selected_greyscale_light900_1I-H1 MenuTabTemplate_item_9qrlt'})
comp.append(channel_no['href'])
a_tags=soup.find_all("a",{'class': "VideoBoxLinkWrapper_wrap_GLkZS lico-home__sc-beae65ec-0 kobjgv BroadcastUnderCard_link_video_vOz9X"})
href_link=[]
for a_tag in a_tags:
    tmp=a_tag["href"]
    if 'replay' in tmp: # 링크에 naver가 포함된 경우만 추가
        href_link.append(tmp)
        
comp.append(href_link[-1])
comp.append(href_link[0])
```

```python
#driver.get(comp[1])
#time.sleep(1)
try:
    driver.find_element(By.PARTIAL_LINK_TEXT,"라이브 소개").click()
except:
    print("ok")
    
    
html = driver.page_source
soup = bs(html, 'html.parser')

title = soup.find('strong', {'class':'IntroductionModalTitle_wrap_wNoB1'})
comp.append(title.get_text())
live_date = soup.find('time',{'class':'LiveStatusTime_wrap_Jf2sB'})
date = live_date['datetime'][0:10]
comp.append(date)
date
```

```python
#driver.get(comp[2])
#time.sleep(1)
try:
    driver.find_element(By.PARTIAL_LINK_TEXT,"라이브 소개").click()
except:
    print("ok")

html = driver.page_source
soup = bs(html, 'html.parser')

title = soup.find('strong', {'class':'IntroductionModalTitle_wrap_wNoB1'})
comp.append(title.get_text())
live_date = soup.find('time',{'class':'LiveStatusTime_wrap_Jf2sB'})
date = live_date['datetime'][0:10]
comp.append(date)
channel_product.append(comp)
date
```
